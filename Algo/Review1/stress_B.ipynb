```python
import subprocess
import timeit
from random import choice, seed

SAMPLES_NUM = 10
ALPHA_SIZE = 26
STRLEN = 1000
SAMPLE_LEN = 100

seed()

def testgen(num=SAMPLES_NUM, str_len=STRLEN, alpha=ALPHA_SIZE):
    chars = [chr(ord('a') + i) for i in range(ALPHA_SIZE)]
    with open('B.in', 'w') as ftest:
        ftest.write(str(str_len) + " " + str(num) + " " + str(alpha) + "\n")
        for i in range(num):
            sample = ''
            for k in range(choice(range(SAMPLE_LEN, SAMPLE_LEN + 1))):
                sample += choice(chars)
            ftest.write(sample + '\n')        

def timerun(num=SAMPLES_NUM, str_len=STRLEN, alpha=ALPHA_SIZE, repeats=1):
    testgen(num, str_len, alpha)
    t = %timeit -o -q -n 1 -r $repeats subprocess.check_output(["./a.out"])
    return t.best
```

```python
found = False
while not found:
    t = timerun()
    if t > 0.1:
        print(t)
        found = True
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x = np.linspace(1, 102, 20)
y = [timerun(int(num)) for num in x]
line, = plt.plot(x, y, '--', linewidth=2, color='green')

plt.show()
```
